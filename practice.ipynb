```python
# 1. Յուրաքանչյուր տարի, ինչ է արտահանել  Հնդկաստանը ամենից շատ.

# որպեսզի իմանանք հարցի պատասխանը, մեզ պետք կգա վերբեռնել  export.csv ֆայլը:

df = pd.read_csv("E:\\dataset\\india-trade-data\\2018-2010_export.csv")
df.head()
```
